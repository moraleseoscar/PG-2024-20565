```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ruta base donde se encuentran las carpetas
base_directory = r'Dexketoprofen'
exp_directory = 'Dexketoprofen\DEXK_3.CSV'

df_exp = pd.read_csv(exp_directory, names=['Long', 'Abs'], skiprows=1)


# Funcin para normalizar una columna
def normalize(column):
    return (column - column.min()) / (column.max() - column.min())

# Funcin para procesar cada archivo .dat
def process_dat_file(file_path):
    df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=['CM-1', 'Total', '2.36A', '3.31A', '5.99A'])
    df['NM'] = 10000000 / df['CM-1']
    df['Total'] = df['Total'] / 100000

    df = df.drop(['2.36A', '3.31A', '5.99A'], axis='columns')
    df = df[['CM-1', 'NM', 'Total']]
    
    plt.figure(figsize=(10, 6))
    plt.plot(df['NM'], df['Total'], label='Total Computacional')
    plt.plot(df_exp['Long'], df_exp['Abs'], label='Total Experimental')
    plt.xlabel('NM (nm)')
    plt.ylabel('Intensidad')
    plt.title(f"Espectro para {os.path.basename(file_path)}")
    plt.legend()
    plt.xlim([150, 350])
    plt.grid(True)
    plt.show()

# Usar os.walk para buscar los archivos
found_files = []
for root, dirs, files in os.walk(base_directory):
    for file in files:
        if file.endswith('_toSpect.out.abs.dat'):
            file_path = os.path.join(root, file)
            found_files.append(file_path)
            

# Si no se encuentran archivos, mostrar un mensaje
if not found_files:
    print("No se encontraron archivos .dat que terminen en '_toSpect.out.abs.dat'.")
else:
    for file_path in found_files:
        print(f"Procesando archivo: {file_path}")
        process_dat_file(file_path)
```
